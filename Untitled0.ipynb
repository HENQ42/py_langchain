```python
import textract

doc = textract.process("./FAQ-1.pdf")

with open('FAQ-1.txt', 'w', encoding="utf-8") as f:
    f.write(doc.decode('utf-8'))

with open('FAQ-1.txt', 'r', encoding="utf-8") as f:
    text = f.read()

tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

def count_tokens(text: str) -> int:
    return len(tokenizer.encode(text))

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=512,
    chunk_overlap=24,
    length_function=count_tokens,
)

chunks = text_splitter.create_documents([text])
```

```python
openai_key = "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"  # Replace with your key
embeddings = OpenAIEmbeddings(openai_api_key=openai_key, model="text-embedding-ada-002")
db = FAISS.from_documents(chunks, embeddings)
```

```python
query = "quem eu preciso falar para comprar a infodental"
docs = db.similarity_search(query)
docs[0]
```

```python
chain = load_qa_chain(OpenAI(openai_api_key=openai_key, temperature=0), chain_type="stuff")
response = chain.run(input_documents=docs, question=query)
response
```
